# Seaborn - Categorical Plots

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

### Bar Plot
A bar plot is a plot that presents categorical data with rectangular bars with lengths proportional to the values that they represent.

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

##### Average or Mean is displayed per categorical value

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips, estimator=np.std)
```

##### Estimator - statistical function to estimate within each categorical bin.

### Count Plot
It shows the counts of observations in each categorical bin using bars.

```python
sns.countplot(x='sex',data=tips)
```

### Box Plot
It is a method for graphically depicting groups of numerical data through their quartiles.

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

### Violin Plot
A violin plot is a method of plotting numeric data. It is similar to a box plot, with the addition of a rotated kernel density plot on each side. Violin plots are similar to box plots, except that they also show the probability density of the data at different values, usually smoothed by a kernel density estimator.

```python
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

### Strip Plot
A strip plot is a scatter plot where one of the variables is categorical. They can be combined with other plots to provide additional information.

```python
sns.stripplot(x='day',y='total_bill',data=tips)
```

##### Amount of jitter (half the width of the uniform random variable support), or just use True for a good default.

```python
sns.stripplot(x='day',y='total_bill',data=tips, jitter=False)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips, hue='sex')
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex', dodge=True)
```

### Swarm Plot
A swarm plot is very similar to a strip plot, yet the locations of points are adjusted automatically to avoid overlap even if the jitter value is not applied.

Combination of Strip and Violin plot.

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips, color='white')
```

### Factor Plot

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='bar')
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='bar')
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='violin')
```
