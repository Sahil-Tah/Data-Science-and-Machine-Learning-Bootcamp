# K Means Clustering

### Check out this [link](https://www.analyticsvidhya.com/blog/2019/08/comprehensive-guide-k-means-clustering/) for better understanding of the concept.

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

### Generating artificial data

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8, random_state=101)
```

```python
data[0].shape
```

```python
plt.scatter(data[0][:,0], data[0][:,1])
```

```python
plt.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')
```

### K Means Model

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

### Plotting

```python
fig, (ax1,ax2) = plt.subplots(1,2, sharey=True, figsize=(10,6))

ax1.set_title('K Means, k=4')
ax1.scatter(data[0][:,0], data[0][:,1], c=kmeans.labels_, cmap='rainbow')

ax2.set_title('Original')
ax2.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')
```

#### Note:

This is just to check how the K means performs to the dummy data. In reality, there won't be anything to check or compare with since K means is unsupervised algorithm so we will not have any labeled data.

---------

### Checking the Output for different K values

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(data[0])

fig, (ax1,ax2) = plt.subplots(1,2, sharey=True, figsize=(10,6))

ax1.set_title('K Means, k=2')
ax1.scatter(data[0][:,0], data[0][:,1], c=kmeans.labels_, cmap='rainbow')

ax2.set_title('Original')
ax2.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(data[0])

fig, (ax1,ax2) = plt.subplots(1,2, sharey=True, figsize=(10,6))

ax1.set_title('K Means, k=3')
ax1.scatter(data[0][:,0], data[0][:,1], c=kmeans.labels_, cmap='rainbow')

ax2.set_title('Original')
ax2.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')
```
