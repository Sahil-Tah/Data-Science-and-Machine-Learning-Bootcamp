# Seaborn - Regression Plots

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

### Linear Model Plot

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'])
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],scatter_kws={'s':100})
```

##### We're passing a dictionary with base matplotlib arguments. Here s is size of a scatter plot.

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day',hue='sex')
```

### Aspect and Size Ratio

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day',hue='sex',aspect=0.6,size=8)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day',hue='sex',aspect=0.6,height=8)
```
