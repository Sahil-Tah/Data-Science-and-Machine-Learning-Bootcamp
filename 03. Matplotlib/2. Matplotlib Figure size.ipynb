# Matplotlib - Figure Size, Aspect Ratio and DPI

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
fig = plt.figure(figsize=(3,2))
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
fig = plt.figure(figsize=(8,2), dpi=100)
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
fig, axes = plt.subplots(figsize=(4,4))
axes.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1,figsize=(6,4))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

### Saving a fig

```python
fig.savefig('my_plot.png', dpi=200)
```

### Axes - Legend

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

axes.plot(x,x**2)
axes.plot(x,x**3)
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

axes.plot(x,x**2,label='X squared')
axes.plot(x,x**3,label='X cubed')

axes.legend()
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

axes.plot(x,x**2,label='X squared')
axes.plot(x,x**3,label='X cubed')

axes.legend(loc=1) # default loc=0
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

axes.plot(x,x**2,label='X squared')
axes.plot(x,x**3,label='X cubed')

axes.legend(loc=(0.1,0.6)) # If the options doesn't work, tuple with (x,y) values can be inserted too
```
