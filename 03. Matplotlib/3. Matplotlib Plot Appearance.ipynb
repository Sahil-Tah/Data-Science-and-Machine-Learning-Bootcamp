# Matplotlib - Plot Appearance

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

### Color Parameter

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(x,y, color='red')
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(x,y, color='#FF8C00') # HEX Values
```

### Linewidth

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(x,y, color='red', linewidth=5)
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(x,y, color='red', linewidth=5, alpha=0.5)
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(x,y, color='red', lw=5)
```

##### Linewidth or the shorthand 'LW' works too

### Linestyle

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(x,y, color='red', lw=5, linestyle='--')
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(x,y, color='red', lw=5, ls='-.')
```

##### Shorthand 'LS' works for Linesight 

### Marker

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(x,y, color='red', lw=3, ls='-', marker='o')
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(x,y, color='red', lw=1, ls='-', marker='+', markersize = 20)
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(x,y, color='purple', lw=1, ls='-', marker='o', markersize = 20)
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(x,y, color='purple', lw=1, ls='-', marker='o', markersize = 20, markerfacecolor='yellow', markeredgewidth=3)
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(x,y, color='purple', lw=1, ls='-', marker='o', markersize = 20, markerfacecolor='yellow', 
          markeredgewidth=3, markeredgecolor='green')
```

### Plot Range

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(x,y, color='purple', lw=2, ls='--')

axes.set_xlim([0,1])
axes.set_ylim([0,2])
```
